```python
import pandas as pd
import numpy as np
df1 = pd.read_csv('C:/Users/Hrach/Desktop/DSML/ML Lessons/homework4/hospital_train.csv')
df2 = pd.read_csv('C:/Users/Hrach/Desktop/DSML/ML Lessons/homework4/hospital_test_X.csv')
```

```python
df1
```

```python
df2['Survived_1_year'] = 0
df2
```

```python
df = pd.concat([df1, df2])
```

```python
for i in df.columns:
    if df[i].dtype == 'object':
        df[i] = df[i].str.strip()
    else:
        pass
df.Treated_with_drugs = df.Treated_with_drugs.str.upper()
df.Patient_Smoker = df.Patient_Smoker.str.upper()
df.Patient_Smoker = df.Patient_Smoker.str.replace("YESS", "YES")
df.drop(['ID_Patient_Care_Situation', 'Patient_ID', 'Patient_mental_condition'], axis=1, inplace=True)
```

```python
df['Treated_with_drugs'] = df['Treated_with_drugs'].fillna('DX6')
df.fillna(-999, inplace=True)
```

```python
for col in df.columns:
    if 'picture' not in col:
        print('{} : {}'.format(col, df[col].unique()))
```

```python
df = pd.get_dummies(df, columns=['Diagnosed_Condition', 'Treated_with_drugs', 'Patient_Smoker', 'Patient_Rural_Urban'])
```

```python
df1 = df.iloc[:22571, :]
df2 = df.iloc[22571:, :]
```

```python
df2.drop('Survived_1_year', axis=1, inplace=True)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, f1_score

xTrain, xTest, yTrain, yTest = train_test_split(df1[df1.columns.difference(['Survived_1_year'])], 
                                                df1['Survived_1_year'], test_size=0.25, random_state=42)
```

```python
# Random Forest

from sklearn.ensemble import RandomForestClassifier

clf = RandomForestClassifier(max_depth=30, random_state=30)
clf.fit(xTrain, yTrain)
y_pred = clf.predict(xTest)

print(f1_score(yTest, y_pred))
print(accuracy_score(yTest, y_pred))
```

```python
from sklearn.model_selection import GridSearchCV

param_grid = {'max_leaf_nodes': [5, 10, 20, 50, 100, 150, 300],
              'max_depth': [5, 10, 20, 30, 50],}
grid = GridSearchCV(RandomForestClassifier(), param_grid, refit = True, verbose = 3)
grid.fit(xTrain, yTrain)
```

```python
print(f1_score(grid.best_estimator_.predict(xTest), yTest))
print(accuracy_score(grid.best_estimator_.predict(xTest), yTest))
grid.best_params_
```

```python
# SVC

from sklearn.svm import SVC

svm_classifier = SVC(C=100, gamma=0.00003)
svm_classifier.fit(xTrain, yTrain)
y_pred = svm_classifier.predict(xTest)
print(f1_score(yTest, y_pred))
print(accuracy_score(yTest, y_pred))
```

```python
# Decision Tree

from sklearn.tree import DecisionTreeClassifier
dtc = DecisionTreeClassifier(max_leaf_nodes=110)
dtc.fit(xTrain, yTrain)
print(f1_score(dtc.predict(xTest), yTest))
print(accuracy_score(dtc.predict(xTest), yTest))
```

```python
from xgboost import XGBClassifier

model = XGBClassifier()
model.fit(xTrain, yTrain)
print(f1_score(model.predict(xTest), yTest))
print(accuracy_score(model.predict(xTest), yTest))
```

```python
import lightgbm as lgb

model1 = lgb.LGBMClassifier(learning_rate=0.09,max_depth=-5,random_state=42)
model1.fit(xTrain, yTrain,eval_metric='logloss')
print(f1_score(model1.predict(xTest), yTest))
print(accuracy_score(model1.predict(xTest), yTest))
```

```python
df2 = df2.reindex(sorted(df2.columns), axis=1)
```

```python
predicted_labels = pd.DataFrame(model1.predict(df2))
predicted_labels.rename(columns={0: 'Survived_1_year'}, inplace=True)
```
